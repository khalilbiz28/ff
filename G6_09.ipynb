```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/EMSIMa/ADD3IIR/main/Projet/Datasets/netflix.csv')
```

## **Question 1: Top 5 des meilleures cat√©gories**

On affiche la base de donnees a analyser

```python
df
```

```python
new_data=df.dropna(subset=['type'])

num_cat=new_data.groupby('type').count()

sort_cat=num_cat.sort_values(by=['show_id'])

top5=sort_cat.head(5)
top5
```

**new_data = df.dropna(subset=['type'])**: Cette ligne de code cr√©e un nouveau DataFrame appel√© new_data en supprimant toutes les lignes o√π la colonne 'type' contient des valeurs manquantes (NaN).
 **num_cat = new_data.groupby('type').count()**: Cette ligne de code groupe les donn√©es de new_data par la colonne 'type' et compte le nombre de valeurs non nulles dans chaque groupe.

**top5 = sort_cat.head(5)**: Cette ligne de code s√©lectionne les cinq premi√®res lignes de la s√©rie sort_cat, ce qui correspond aux types ayant le nombre le plus bas d'occurrences. Ces cinq types sont assign√©s au DataFrame top5.

```python
category_counts = df['type'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title("R√©partition des cat√©gories de produits")
plt.xlabel("Cat√©gorie")
plt.ylabel("Nombre de produits")
plt.show()
```

## Question 2: Top 5 des r√©alisateurs

## Question 4 : Les 10 meilleurs acteurs

```python
top_actors=df['cast'].str.split(',', expand=True).stack().value_counts().head(10)
top_actors
```

**df['cast'].str.split(',', expand=True):** Ce code divise les valeurs de la colonne 'cast' du DataBase df en utilisant la virgule comme d√©limiteur. L'argument expand=True permet d'√©tendre le r√©sultat en plusieurs colonnes, o√π chaque colonne repr√©sente un acteur. Les valeurs de la colonne 'cast' sont s√©par√©es en fonction des virgules et r√©parties en colonnes s√©par√©es.

**stack():** Cette m√©thode transforme les colonnes r√©sultantes de l'√©tape pr√©c√©dente en une s√©rie. Chaque valeur d'acteur devient une entr√©e distincte dans la s√©rie, tandis que les indices de la s√©rie contiennent des informations sur les colonnes et les lignes d'origine.

**value_counts():** Cette m√©thode compte le nombre d'occurrences de chaque acteur dans la s√©rie r√©sultante. Elle renvoie une nouvelle s√©rie o√π les indices sont les noms des acteurs et les valeurs correspondent au nombre d'occurrences de chaque acteur dans la s√©rie.

**head(10):** Cette m√©thode s√©lectionne les dix premi√®res lignes de la s√©rie r√©sultante, ce qui correspond aux dix acteurs les plus fr√©quents. Ces dix acteurs sont assign√©s √† la variable top_actors.

## Question 5: Les cinq meilleures s√©ries t√©l√©vis√©es avec le plus grand nombre de saisons

```python
series_tv = df[df['type'] == 'TV Show']

top_series = series_tv.sort_values(by='duration', ascending=False).head(20)

print(top_series[['title', 'duration']])
```

**series_tv = df[df['type'] == 'TV Show']**: Cette ligne filtre les donn√©es de la base de donn√©es df pour ne conserver que les entr√©es correspondant aux s√©ries t√©l√©vis√©es. Elle cr√©e un nouveau DataFrame appel√© series_tv qui contient uniquement les lignes o√π la valeur de la colonne "type" est √©gale √† "TV Show".

**top_series = series_tv.sort_values(by='duration', ascending=False).head(20)**: Cette ligne trie le DataFrame series_tv par ordre d√©croissant en fonction de la colonne "duration" qui repr√©sente la dur√©e de la s√©rie. La fonction sort_values() effectue le tri, avec l'argument by='duration' pour sp√©cifier la colonne utilis√©e pour le tri. L'argument ascending=False indique que le tri doit √™tre effectu√© dans l'ordre d√©croissant. Ensuite, la m√©thode head(20) s√©lectionne les 20 premi√®res lignes du DataFrame tri√© et les assigne au DataFrame top_series.

**print(top_series[['title', 'duration']]):** Cette ligne affiche les colonnes "title" (titre de la s√©rie) et "duration" (dur√©e de la s√©rie) du DataFrame top_series. Cela permet d'afficher le titre et la dur√©e des 20 meilleures s√©ries t√©l√©vis√©es avec le plus grand nombre de saisons.

## Question 9: Affichage des les dix pays les plus fr√©quents en fonction de l'ann√©e de sortie pour chaque pays

```python
countries = df['country'].value_counts().head(10).index # Choix des 10 pays les plus fr√©quents
x = df[df['country'].isin(countries)]['country']
y = df[df['country'].isin(countries)]['release_year'].astype(int)

plt.scatter(x, y)
plt.xlabel('Pays')
plt.ylabel('Ann√©e de sortie')
plt.title('Ann√©es de sortie par pays')
plt.xticks(rotation=60)
plt.show()
```

**countries = df['country'].value_counts().head(10).index** : Cecode compte le nombre d'occurrences de chaque pays dans la colonne 'country' et s√©lectionne les dix pays les plus fr√©quents. Les noms des pays sont extraits en tant qu'index de la s√©rie r√©sultante et assign√©s √† la variable countries.

**x = df[df['country'].isin(countries)]['country']** : Cette ligne de code filtre le DataFrame df pour ne conserver que les lignes o√π la colonne 'country' contient un des dix pays s√©lectionn√©s dans la variable countries. Elle extrait ensuite ces noms de pays dans la variable x.

**y = df[df['country'].isin(countries)]['release_year'].astype(int**) : Cette ligne effectue un filtrage similaire au point pr√©c√©dent, mais cette fois pour la colonne 'release_year' (ann√©e de sortie). Elle extrait les ann√©es de sortie correspondantes dans la variable y et les convertit en type entier √† l'aide de la m√©thode astype(int).

**plt.scatter(x, y) :** Ce code cr√©e un graphe de dispersion en utilisant les valeurs de x (pays) sur l'axe horizontal et les valeurs de y (ann√©es de sortie) sur l'axe vertical. Chaque point repr√©sente un film et sa position est d√©termin√©e par le pays correspondant et l'ann√©e de sortie.

**plt.xlabel('Pays')**: Cette ligne de code d√©fini le label de l'axe horizontal (l'axe des x) en tant que "Pays".

**plt.ylabel('Ann√©e de sortie')** : Cette ligne de code d√©fini le label de l'axe vertical (l'axe des y) en tant que "Ann√©e de sortie".

**plt.title('Ann√©es de sortie par pays')** : le titre du graphe en tant que "Ann√©es de sortie par pays".

**plt.xticks(rotation=60)** : Ce code fait pivoter les √©tiquettes des pays sur l'axe horizontal de 60 degr√©s afin de les rendre plus lisibles.

**plt.show()** : Cette ligne de code affiche le graphe de dispersion √† l'√©cran.
