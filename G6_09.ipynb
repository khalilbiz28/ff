```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/EMSIMa/ADD3IIR/main/Projet/Datasets/netflix.csv')
```

## **Question 1: Top 5 des meilleures catégories**

On affiche la base de donnees a analyser

```python
df
```

```python
new_data=df.dropna(subset=['type'])

num_cat=new_data.groupby('type').count()

sort_cat=num_cat.sort_values(by=['show_id'])

top5=sort_cat.head(5)
top5
```

**new_data = df.dropna(subset=['type'])**: Cette ligne de code crée un nouveau DataFrame appelé new_data en supprimant toutes les lignes où la colonne 'type' contient des valeurs manquantes (NaN).
 **num_cat = new_data.groupby('type').count()**: Cette ligne de code groupe les données de new_data par la colonne 'type' et compte le nombre de valeurs non nulles dans chaque groupe.

**top5 = sort_cat.head(5)**: Cette ligne de code sélectionne les cinq premières lignes de la série sort_cat, ce qui correspond aux types ayant le nombre le plus bas d'occurrences. Ces cinq types sont assignés au DataFrame top5.

```python
category_counts = df['type'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title("Répartition des catégories de produits")
plt.xlabel("Catégorie")
plt.ylabel("Nombre de produits")
plt.show()
```

## Question 4 : Les 10 meilleurs acteurs

```python
top_actors=df['cast'].str.split(',', expand=True).stack().value_counts().head(10)
top_actors
```

**df['cast'].str.split(',', expand=True):** Ce code divise les valeurs de la colonne 'cast' du DataBase df en utilisant la virgule comme délimiteur. L'argument expand=True permet d'étendre le résultat en plusieurs colonnes, où chaque colonne représente un acteur. Les valeurs de la colonne 'cast' sont séparées en fonction des virgules et réparties en colonnes séparées.

**stack():** Cette méthode transforme les colonnes résultantes de l'étape précédente en une série. Chaque valeur d'acteur devient une entrée distincte dans la série, tandis que les indices de la série contiennent des informations sur les colonnes et les lignes d'origine.

**value_counts():** Cette méthode compte le nombre d'occurrences de chaque acteur dans la série résultante. Elle renvoie une nouvelle série où les indices sont les noms des acteurs et les valeurs correspondent au nombre d'occurrences de chaque acteur dans la série.

**head(10):** Cette méthode sélectionne les dix premières lignes de la série résultante, ce qui correspond aux dix acteurs les plus fréquents. Ces dix acteurs sont assignés à la variable top_actors.

## Question 5: Les cinq meilleures séries télévisées avec le plus grand nombre de saisons

```python
series_tv = df[df['type'] == 'TV Show']

top_series = series_tv.sort_values(by='duration', ascending=False).head(20)

print(top_series[['title', 'duration']])
```

**series_tv = df[df['type'] == 'TV Show']**: Cette ligne filtre les données de la base de données df pour ne conserver que les entrées correspondant aux séries télévisées. Elle crée un nouveau DataFrame appelé series_tv qui contient uniquement les lignes où la valeur de la colonne "type" est égale à "TV Show".

**top_series = series_tv.sort_values(by='duration', ascending=False).head(20)**: Cette ligne trie le DataFrame series_tv par ordre décroissant en fonction de la colonne "duration" qui représente la durée de la série. La fonction sort_values() effectue le tri, avec l'argument by='duration' pour spécifier la colonne utilisée pour le tri. L'argument ascending=False indique que le tri doit être effectué dans l'ordre décroissant. Ensuite, la méthode head(20) sélectionne les 20 premières lignes du DataFrame trié et les assigne au DataFrame top_series.

**print(top_series[['title', 'duration']]):** Cette ligne affiche les colonnes "title" (titre de la série) et "duration" (durée de la série) du DataFrame top_series. Cela permet d'afficher le titre et la durée des 20 meilleures séries télévisées avec le plus grand nombre de saisons.

## Question 7: Netflix se concentre-t-il davantage sur les séries télévisées que sur les films ces dernières années ?

```python
netflix_movies = df[(df["type"] == "Movie") & (df["release_year"] >= 2019)]


netflix_tv_shows = df[(df["type"] == "TV Show") & (df["release_year"] >= 2019)]


print("Netflix -- { Nombre de films : ", len(netflix_movies), " | Nombre de séries télévisées : ", len(netflix_tv_shows))

```

Ce code permet de filtrer le DataFrame df pour sélectionner uniquement les films et les séries télévisées sortis en 2019 ou après sur Amazon Prime Video. Il affiche ensuite le nombre de films et de séries télévisées correspondants.

**amazon_movies = df[(df["type"] == "Movie") & (df["release_year"] >= 2019)] **: Dans cette ligne on crée une nouvelle DataBase appelé amazon_movies en filtrant la DataBase df. Il ne conserve que les lignes où la colonne "type" est "Movie" (film) et où la colonne "release_year" est supérieure ou égale à 2019.

**amazon_tv_shows = df[(df["type"] == "TV Show") & (df["release_year"] >= 2019)]** : Dans cette ligne on crée une nouvelle DataBase appelé amazon_tv_shows en filtrant la DataBase df. Il ne conserve que les lignes où la colonne "type" est "TV Show" (série télévisée) et où la colonne "release_year" est supérieure ou égale à 2019.

**print("Amazon Prime Video -- { Nombre de films : ", len(amazon_movies), " | Nombre de séries télévisées : ", len(amazon_tv_shows)) **: Ce code affiche le nombre de films et de séries télévisées disponibles sur Amazon Prime Video. Elle utilise la fonction len() pour compter le nombre de lignes dans les DataFrames amazon_movies et amazon_tv_shows.

## Question 9: Affichage des les dix pays les plus fréquents en fonction de l'année de sortie pour chaque pays

```python
countries = df['country'].value_counts().head(10).index # Choix des 10 pays les plus fréquents
x = df[df['country'].isin(countries)]['country']
y = df[df['country'].isin(countries)]['release_year'].astype(int)

plt.scatter(x, y)
plt.xlabel('Pays')
plt.ylabel('Année de sortie')
plt.title('Années de sortie par pays')
plt.xticks(rotation=60)
plt.show()
```

**countries = df['country'].value_counts().head(10).index** : Cecode compte le nombre d'occurrences de chaque pays dans la colonne 'country' et sélectionne les dix pays les plus fréquents. Les noms des pays sont extraits en tant qu'index de la série résultante et assignés à la variable countries.

**x = df[df['country'].isin(countries)]['country']** : Cette ligne de code filtre le DataFrame df pour ne conserver que les lignes où la colonne 'country' contient un des dix pays sélectionnés dans la variable countries. Elle extrait ensuite ces noms de pays dans la variable x.

**y = df[df['country'].isin(countries)]['release_year'].astype(int**) : Cette ligne effectue un filtrage similaire au point précédent, mais cette fois pour la colonne 'release_year' (année de sortie). Elle extrait les années de sortie correspondantes dans la variable y et les convertit en type entier à l'aide de la méthode astype(int).

**plt.scatter(x, y) :** Ce code crée un graphe de dispersion en utilisant les valeurs de x (pays) sur l'axe horizontal et les valeurs de y (années de sortie) sur l'axe vertical. Chaque point représente un film et sa position est déterminée par le pays correspondant et l'année de sortie.

**plt.xlabel('Pays')**: Cette ligne de code défini le label de l'axe horizontal (l'axe des x) en tant que "Pays".

**plt.ylabel('Année de sortie')** : Cette ligne de code défini le label de l'axe vertical (l'axe des y) en tant que "Année de sortie".

**plt.title('Années de sortie par pays')** : le titre du graphe en tant que "Années de sortie par pays".

**plt.xticks(rotation=60)** : Ce code fait pivoter les étiquettes des pays sur l'axe horizontal de 60 degrés afin de les rendre plus lisibles.

**plt.show()** : Cette ligne de code affiche le graphe de dispersion à l'écran.
